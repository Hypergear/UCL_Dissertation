```python
def get_file_path(root_path):
    file_list = []
    dir_list = []
    #获取该目录下所有的文件名称和目录名称
    dir_or_files = os.listdir(root_path)
    for dir_file in dir_or_files:
        #获取目录或者文件的路径
        dir_file_path = os.path.join(root_path,dir_file)
        #判断该路径为文件还是路径
        if os.path.isdir(dir_file_path):
            dir_list.append(dir_file_path)
            #递归获取所有文件和目录的路径
            get_file_path(dir_file_path,file_list,dir_list)
        else:
            file_list.append(dir_file_path)
    return [file_list,dir_list]

def filter_unnecessary_files(list_cable_fileName,cable_path):
    list_cable_fileName[0] =[val  for val in list_cable_fileName[0] if re.search(r'\w*(.json)$', val)]
    # Some files are not needed in the cable folder
    list_notneeded_file = ['cable-geo.json', 'all.json']
    # list_notneeded_file = [os.path.join(root_path,x) for x in list_notneeded_file]   # Explains the next line
    list_cable_fileName[0] = [x for x in list_cable_fileName[0] if x not in [os.path.join(cable_path,x) for x in list_notneeded_file]]
    return list_cable_fileName

def read_cable_json(file_list):
    df_nested_list = pd.DataFrame()
    for json_file in file_list:
        with open(json_file,'r') as f:
            cable_data = json.loads(f.read())
        temp = pd.read_json(json_file).rename(columns={'id':'cable_id','name':'cable_name','length':'cable_length','owners':'cable_owners','suppliers':'cable_suppliers','rfs':'cable_finish','rfs_year':'cable_finish_year','is_planned':'cable_is_planned','url':'cable_url','notes':'cable_notes'}).join(pd.json_normalize(cable_data, record_path =['landing_points']).rename(columns={'id': 'landingPoint_id', 'name': 'landingPoint_name', 'country': 'landingPoint_country', 'is_tbd': 'landingPoint_is_tbd'})).drop(['landing_points'],axis=1)
        df_nested_list = df_nested_list.append(temp,ignore_index=True)
    return df_nested_list

# todo
def read_landingPoint_geojson(landingPoint_path):
    with open(landingPoint_path,'r') as f:
        landingPoint_data = json.loads(f.read())
    pd.read_json
    return None

# todo
def merge_cable_with_geojson():
    return None

cable_path = '/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/cable/'
landingPoint_path = '/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/landing-point/landing-point-geo.json'

list_cable_fileName = get_file_path(cable_path)
list_cable_fileName = filter_unnecessary_files(list_cable_fileName,cable_path)



df_all_cables = read_cable_json(list_cable_fileName[0])
df_all_cables.index.name = 'id'
# len(list_cable_fileName[0])

# list_cable_fileName[0]

# print(read_from_json(['/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/cable/colombian-festoon.json','/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/cable/alaska-united-east-au-east.json']).tail())

# df_all_cables.to_csv('/Users/dijiongyuequ/Downloads/tem.csv',header=0)

# with open('/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/cable/2africa.json','r') as f:
#     data = json.loads(f.read())
#     print(type(pd.read_json('/Users/dijiongyuequ/Downloads/GitHub/UCL_Dessertation/Dataset/SubmarineCableMap/v3/cable/2africa.json').join(pd.json_normalize(data, record_path =['landing_points']).rename(columns={'id': 'landing_id', 'name': 'landing_name', 'country': 'landing_country', 'is_tbd': 'landing_is_tbd'})).drop(['landing_points'],axis=1)))

```
